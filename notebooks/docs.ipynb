## Módulo de autenticación

El archivo ```auth_controller.py``` contiene el endpoint ```/login/email/nombres/apellido/dni``` que recibe las solicitudes de inicio de sesión provenientes de **ECONET**:

````python
from services.usuario_service import UsuarioService
from config.db_config import get_db
from sqlalchemy.ext.asyncio import AsyncSession
from fastapi import Depends, APIRouter

router = APIRouter(
    prefix="/auth",
    tags=["Auth"]
)

def get_usuario_service(db: AsyncSession = Depends(get_db)) -> UsuarioService:
    return UsuarioService(db)

@router.post("/login/{email}/{dni}/{nombres}/{apellido}")
async def login (
        email: str,
        dni: int,
        nombres: str,
        apellido: str,
        svc: UsuarioService = Depends(get_usuario_service)
) -> dict:
    return await svc.check_login_or_register(email=email, nombres=nombres, apellido=apellido, dni=dni)
````

El método ```check_login_or_register``` se va a encargar de comprobar si el usuario existe o no. Si el usuario no existe, sus datos se van a enviar al método ```register``` de ```auth_service.py```. Si existe, sus datos se van a enviar a ```iniciar_sesion_econet```, ubicado en el mismo archivo.
